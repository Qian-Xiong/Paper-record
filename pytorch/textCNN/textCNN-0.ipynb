```python
import re

from tqdm import tqdm
import os
import jieba

#定义参数
max_len=30 ##最大长度
UNK,PAD="UNK","PAD"

##获取数据，使用数据集THUNews
train_content=[]
train_label=[]
with open("./data/THUCNews/test.txt","r",encoding="utf-8") as f:
    for line in tqdm(f):#进度条
        line=line.strip()
        # print(line,end="\n\n")
        if len(line)>3:
            content=line.split("\t")[0][:max_len]
            train_content.append(content)
            train_label.append(line.split("\t")[1])
print(len(train_content))
print(len(train_label))
print(train_content)

```

```python
#将数据分词，并转化为索引
for i in range(len(train_content)):
    train_content[i]=re.sub('\s','',train_content[i])
    print(train_content[i])
    out=jieba.cut(train_content[i],cut_all=False)
    train_content[i]="/".join(out).strip().split("/")
    print(train_content[i])

#使用列表去不重提取关键字
dictL=list()
for i in train_content:
    for j in i:
        dictL.append(j)

#使用集合去重提取关键字
dictS=set()
for i in train_content:
    for j in i:
        dictS.add(j)

```

```python
print(dictL)
print(dictS)
```

```python
#显示词频
from collections import Counter
wordcount = Counter(dictL)
print(wordcount)
```
